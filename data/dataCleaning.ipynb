### Fixes:
- DAYOFSERVICE should be a datetime value
- LASTUPDATE should be a datetime value

```python
df['DAYOFSERVICE'] = pd.to_datetime(df['DAYOFSERVICE'])
df['LASTUPDATE'] = pd.to_datetime(df['LASTUPDATE'])
```

### Other changes:
- Need to convert times in seconds to time stamps
- https://www.kite.com/python/answers/how-to-convert-seconds-to-hours,-minutes,-and-seconds-in-python
- convert seconds to hours, minutes, seconds
- converting PLANNEDTIME_ARR, PLANNEDTIME_DEP, ACTUALTIME_ARR, ACTUALTIME_DEP

```python
df['PLANNEDTIME_ARR'] = pd.to_timedelta(df['PLANNEDTIME_ARR'], unit='s')
df['PLANNEDTIME_DEP'] = pd.to_timedelta(df['PLANNEDTIME_DEP'], unit='s')
df['ACTUALTIME_ARR'] = pd.to_timedelta(df['ACTUALTIME_ARR'], unit='s')
df['ACTUALTIME_DEP'] = pd.to_timedelta(df['ACTUALTIME_DEP'], unit='s')
```

### Convert each of these to datetime next
- We do this by taking the date of the bus journey and add it to the date of the DAYOFSERVICE column

```python
df.loc[0]['DAYOFSERVICE'] + df.loc[0]['PLANNEDTIME_ARR']
```

### Make a new column for each of PLANNEDTIME_ARR, PLANNEDTIME_DEP, ACTUALTIME_ARR, ACTUALTIME_DEP

```python
# make a series for each new column

plannedArrival = []
plannedDeparture = []
actualArrival = []
actualDeparture = []
for i in range(df.shape[0]):
    plannedArrival.append(df.loc[i]['DAYOFSERVICE'] + df.loc[i]['PLANNEDTIME_ARR'])
    plannedDeparture.append(df.loc[i]['DAYOFSERVICE'] + df.loc[i]['PLANNEDTIME_DEP'])
    actualArrival.append(df.loc[i]['DAYOFSERVICE'] + df.loc[i]['ACTUALTIME_ARR'])
    actualDeparture.append(df.loc[i]['DAYOFSERVICE'] + df.loc[i]['ACTUALTIME_DEP'])

plannedArrival = pd.Series(plannedArrival)
plannedDeparture = pd.Series(plannedDeparture)
actualArrival = pd.Series(actualArrival)
actualDeparture = pd.Series(actualDeparture)
```

### Add each new series to the dataframe

```python
df['plannedArrival'] = plannedArrival
df['plannedDeparture'] = plannedDeparture
df['actualArrival'] = actualArrival
df['actualDeparture'] = actualDeparture
```

### Drop the old columns

```python
df.drop(columns='PLANNEDTIME_ARR', inplace=True)
df.drop(columns='PLANNEDTIME_DEP', inplace=True)
df.drop(columns='ACTUALTIME_ARR', inplace=True)
df.drop(columns='ACTUALTIME_DEP', inplace=True)
```
